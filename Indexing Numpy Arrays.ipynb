```
import numpy as np
x = np.arange(2,20,2)
print(x)
```

```
print(x[0])
print(x[1])
print(x[2])
```

```
print(x.size)  #number of elements
```

```
#  array[ start(including) : stop(excluding) : step ]

print(x[0:9:1])
print(x[0:8:1])
print(x[2:9:1])
print(x[0:9:2])
```

- **Filling certain indexes with desired value**

```
y = np.arange(2,22,2)
print(y)
```

```
y[0:10:2]=0
print(y)
```

```
y[1:10:2]=1
print(y)
```

```
y[:3]=4   
print(y)
```

- **Advanced indexing**

Working logic : array_name[row_indexes , column_indexes]

```
array_1 = [[1,2],
           [3,4]]

z = np.array(array_1)
print(z)
```

```
print(z[0,0])
```

```
print(z[0,1])
```

```
print(z[1,0])
```

```
print(z[1,1])
```

```
print(z[[0,0],[0,1]])
```

```
print(z[[1,1],[0,1]])
```
