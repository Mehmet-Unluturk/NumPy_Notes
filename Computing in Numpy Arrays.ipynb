- **Arithmetic Operations with Two Arrays**

```
import numpy as np 
liste_1 = [1,3,5,7]
liste_2 = [2,4,6,8]
dizi_1 = np.array(liste_1)
dizi_2 = np.array(liste_2)
print(dizi_1,dizi_2)
```

```
print(dizi_1 + dizi_2)
```

```
print(dizi_1 * dizi_2)
```

- **Comparison Operators**

```
import numpy as np
y = np.arange(1,7)
print(y)
```

```
print(y>3)
```

```
print(y>=3)
```

```
print(y!=3)
```

```
print(y==3)
```

-**Logical Operators**

```
#1- np.logical_and

import numpy as np
print(np.logical_and(True,True))
print(np.logical_and(True,False))
print(np.logical_and(False,False))
```

```
array_1 = np.arange(2,15,3)
print(array_1)
```

```
print(np.logical_and(array_1>2,array_1<11))
```

```
print(np.logical_and(array_1>=5,array_1<13))
```

```
#2- np.logical_or

print(np.logical_or(True,True))
print(np.logical_or(True,False))
print(np.logical_or(False,False))
```

```
print(array_1)
```

```
print(np.logical_or(array_1>2,array_1<11))
```

```
print(np.logical_or(array_1>=5,array_1<13))
```
